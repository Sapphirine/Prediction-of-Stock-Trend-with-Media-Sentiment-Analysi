```python
posi=0
nega=0
for i in range (0,len(ccdata)):
    get_val=ccdata.Negative[i]
    if(float(get_val)<(0.077)):######## use neg as threshold
        nega=nega+1
    if(float(get_val>=(0.077))):
        posi=posi+1
posper=(posi/(len(ccdata)))*100
negper=(nega/(len(ccdata)))*100
print("% of positive tweets= ",posper)
print("% of negative tweets= ",negper)
arr=np.asarray([posper,negper], dtype=int)
mlpt.pie(arr,labels=['positive','negative'])
mlpt.plot()
```

```python
# define train and test set
df_=ccdata[['Date','Comp','Negative','Neutral','Positive','Prices','Prices2','Prices3','Prices4','NASDAQ','NASDAQ2','NASDAQ3','NASDAQ4']].copy()
# df_
train_start_index = 80
train_end_index = 165
test_start_index = 166
test_end_index = 194

# train_start_index = 51
# train_end_index = 194
# test_start_index = 0
# test_end_index = 50

train = df_[train_start_index:train_end_index]
test = df_[test_start_index:test_end_index]
train
```

```python
sentiment_score_list = []

for date, row in train.T.iteritems():

 
    if date>0:
        sentiment_score = np.asarray([df_.loc[date, 'Negative'],df_.loc[date, 'Neutral'],
                                      df_.loc[date, 'Positive'],
                                      df_.loc[date-1, 'Prices'], df_.loc[date, 'NASDAQ'],
                                      df_.loc[date-1,'Prices2'], df_.loc[date, 'NASDAQ2'],
                                      df_.loc[date-1,'Prices3'], df_.loc[date, 'NASDAQ3'],
                                      df_.loc[date-1,'Prices4'], df_.loc[date, 'NASDAQ4']])
        
    else:
        sentiment_score = np.asarray([df_.loc[date, 'Negative'],df_.loc[date, 'Neutral'],
                                      df_.loc[date, 'Positive'],
                                      df_.loc[date, 'Prices'], df_.loc[date, 'NASDAQ'],
                                      df_.loc[date,'Prices2'], df_.loc[date, 'NASDAQ2'],
                                      df_.loc[date,'Prices3'], df_.loc[date, 'NASDAQ3'],
                                      df_.loc[date,'Prices4'], df_.loc[date, 'NASDAQ4']])
 

    sentiment_score_list.append(sentiment_score)

numpy_df_train = np.asarray(sentiment_score_list)
print(numpy_df_train)
print(len(numpy_df_train))
```
